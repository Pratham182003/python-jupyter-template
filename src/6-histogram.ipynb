### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import matplotlib.pyplot as plt
import numpy as np

# Create a figure with 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Data for line plot
x = np.arange(0, 11, 1)
y = x**2
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
axs[0, 0].set_title('Line Plot')
axs[0, 0].set_xlabel('X values')
axs[0, 0].set_ylabel('Y = X^2')
axs[0, 0].grid(True, linestyle='--', alpha=0.7)

# Data for bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Data for scatter plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axs[1, 0].scatter(x_scatter, y_scatter, color='blue')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X values')
axs[1, 0].set_ylabel('Y values')

# Data for pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'green', 'red', 'purple'])
axs[1, 1].set_title('Pie Chart')

# Data for histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
plt.figure(figsize=(6, 4))
plt.hist(data, bins=5, color='cyan', edgecolor='black')
plt.title('Histogram of Data')
plt.xlabel('Data Values')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
